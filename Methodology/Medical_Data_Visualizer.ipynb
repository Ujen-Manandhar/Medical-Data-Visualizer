In this project, you will visualize and make calculations from medical examination data using matplotlib, seaborn, and pandas. The dataset values were collected during medical examinations.

```python
df = pd.read_csv('medical_examination.csv', index_col=0)
```

```python
df.head()
```

Add an overweight column to the data. To determine if a person is overweight,  the person first calculate their BMI by dividing their weight in kilograms by the square of their height in meters. If that value is > 25 then is overweight. Use the value 0 for NOT overweight and the value 1 for overweight.

```python
def bmi(row):
    if (row.weight / (row.height*0.01)**2) > 25:
        return 1
    return 0
```

```python
df['overweight'] = df.apply(bmi, axis=1)
```

```python
df
```

Normalize data by making 0 always good and 1 always bad. If the value of cholesterol or gluc is 1, set the value to 0. If the value is more than 1, set the value to 1.

```python
df['cholesterol'] = df.cholesterol.map(lambda x : 0 if x == 1 else 1)
df['gluc'] = df.gluc.map(lambda x : 0 if x == 1 else 1)
```
